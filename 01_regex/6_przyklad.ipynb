<u>Notatka</u>

W przypadku danych ustrukturyzowanych stosowanie wyrażeń regularnych do parsowania jest przeważnie zbędne. W oparciu o samą struktrę danych najczęściej jest już dość łatwo wyciągnąć odpowiednie dane.
